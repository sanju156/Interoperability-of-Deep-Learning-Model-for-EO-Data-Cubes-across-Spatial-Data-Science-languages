```python
# For Python download the hdf5 file data from the source using the Breizhcrops package (BREIZHCROPS: A TIME SERIES DATASET FOR CROP TYPE MAPPING (2020) - https://doi.org/10.5194/isprs-archives-XLIII-B2-2020-1545-2020)
dataset = BreizhCrops(
    region="frh03",   #(frh01,frh02,frh03,frh04) download all the regions
    root="./breizh_data",
    level="L1C",             
    preload_ram = False
)
```

```python
# For Julia convert the hdf5 files to numpy array
DATA_PATH = "/breizh_data"                  
LEVEL = "L1C"
PRELOAD_RAM = False
frh01 = BreizhCrops(region="frh01", root=DATA_PATH, level=LEVEL, preload_ram=PRELOAD_RAM) #(frh01,frh02,frh03,frh04)
import numpy as np

X = []
y = []
fids = []
for i in range(len(frh01)):
    x, label, fid = frh01[i]
    X.append(x.unsqueeze(0))
    y.append(label)
    fids.append(fid)

X = torch.cat(X).numpy()

np.save("X_frh01.npy", X)
np.save("y_true_frh01.npy", np.array(y))
np.save("field_ids_frh01.npy", np.array(fids))
```
